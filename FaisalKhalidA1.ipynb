```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv('automobile.csv',encoding='unicode_escape')
```

```python
df = df.drop('model_year', axis=1)
df = df.drop('origin', axis=1)
```

```python
df['mpg']=df['mpg'].astype('int')
df['displacement']=df['displacement'].astype('int')
df['horsepower']=df['horsepower'].astype('int')
df['acceleration']=df['acceleration'].astype('int')
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns={'name':'Car Name'},inplace=True)
df.rename(columns={'mpg':'Mpg'},inplace=True)
df.rename(columns={'cylinders':'Cylinders'},inplace=True)
df.rename(columns={'displacement':'Displacement'},inplace=True)
df.rename(columns={'horsepower':'Horsepower'},inplace=True)
df.rename(columns={'weight':'Weight'},inplace=True)
df.rename(columns={'acceleration':'Acceleration'},inplace=True)

```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Cylinders', y='Mpg', data=df)
plt.title("Car Fuel efficiency measured in Miles Per Gallon with Car Cylinders")
plt.xlabel("Number of Cylinders")
plt.ylabel("MPG")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.regplot(x='Horsepower', y='Mpg', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'yellow'})
plt.title("Car Fuel efficiency measured in Miles Per Gallon with Car Horsepower")
plt.xlabel("Horsepower(Power output of the engine)")
plt.ylabel("MPG")
plt.show()
```

```python
plt.figure(dpi=120)
sns.boxplot(x='Acceleration', y='Mpg',data=df)
plt.title("car Fuel efficiency measured in Miles Per Gallon with Acceleration of car")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heatmap for Comparison")
plt.show()
```
